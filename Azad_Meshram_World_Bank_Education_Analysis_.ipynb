```python
# As 3666 indicators are studied by World Bank, we would only require the desired indictors,on which the work has to be done.
''' The Dataframe for the desired 'Indicators' would be helpfull to straight forwardly extract the country specific trends , 
    which are demanded. So our analysis would be restricted to the interlinkages of the given indicators.'''

indicator_df = {'Indicator Name':['Pupil-teacher ratio in secondary education (headcount basis)',
                                  'Adult illiterate population, 15+ years, both sexes (number)',
                                  'Adult literacy rate, population 15+ years, both sexes (%)',
                                  'Unemployment, total (% of total labor force)',
                                  'Government expenditure on education as % of GDP (%)',
                                  'Gross enrolment ratio, secondary, both sexes (%)',
                                  'Lower secondary completion rate, both sexes (%)',
                                  'Labor force, total','Population, total'],
                'Indicator Code' :['SE.SEC.ENRL.TC.ZS','UIS.LP.AG15T99',
                                   'SE.ADT.LITR.ZS','SL.UEM.TOTL.ZS',
                                   'SE.XPD.TOTL.GD.ZS','SE.SEC.ENRR',
                                   'SE.SEC.CMPT.LO.ZS','SL.TLF.TOTL.IN',
                                   'SP.POP.TOTL']}

pd.DataFrame(indicator_df)
```

```python
# Write your code to make your dataset analysis ready.

''' for any kind of country-specific indicator analysis we would require to extract
    the values with respect to the specific country and indicator'''


def Data_Extracter( Country , Indicator_name):

  indi_code= []
  for first in first_df['Indicator Name']:       # This code wil give Indicator code , respected to the indicator name.
    if first == Indicator_name :
      Z= first_df[first_df['Indicator Name']==first]
      Y= Z['Indicator Code'].reset_index()
      indi_code.append(Y.iloc[0])     # as condition would true for 3666 entries but we want only one , so iloc would be useful.
      break

  A = first_df[first_df['Country Name']== Country]  # In first line we would specify the country
  B = A[A['Indicator Code']==indi_code[0][1]]        # In this, we are specifying, the indicator via the variable indi_code.
  B = B.fillna(0).reset_index()     # as the dataset is full of null values,removing them would affect the trend comparison with other indicators.
                                    
  C = B.iloc[0,5:52]                # World bank have filled the values upto 2100, which is not possible as we are in 2023 only. 
                                    # Hence slicing by 'iloc' would help to get the data upto 2015.
  D = pd.DataFrame(C)               # making the individual dataframe for any opertaions.
  D = D.rename(columns={0:Indicator_name})  # assigning the correct name for its column.
  return D
```

```python
# Finally when the indicator trends have been extracted, they need to be shown in one df. So thanks to concatenation.
def Concat(*variable):
  A = pd.concat([*variable],axis=1)                # I want concatenation column wise , so axis= 1
  # A = A.rename(columns={'index':'Years'})        #Years column has to be shown while plotting, hence renaming is important
  return A
```

```python
def Unstacker(*indicator):
  A = pd.concat([*indicator],axis=1)
  B= A.unstack()        # unstacking will make ploting of df very easy and allows to  create colourful variation.
  C= B.reset_index()
  D=C.rename(columns={'level_0':'Countries',0:'Values','level_1':'Years'})  # rename according to the nature of data.
  return D
```

```python
def Present_values(df , Country):
  A= df.replace(0,np.nan)       # again considering the null values is odd.
  B= A.unstack().reset_index()  # But the presence of zero would affect the mean of given column.
  B = B.rename(columns={'level_0':'Indicators',0:Country})  # Hence putting back null values and removing zeoros
  C= B.groupby('Indicators')[Country].mean()  # finally mean of the unique indicators to show the present context of given indicators.
  return C
```

```python
#for Botswana
BOTS_four= Present_values(BOTS_df,'Botswana')

#for East Asia Pacific
EAP_four = Present_values(EAP_df, ' East_Asia_Pacific')

#for Japan
Japan_four = Present_values(Japan_df, 'Japan')

#for China 
China_four = Present_values(china_df, 'China')

#for United_Kingdom 
Uk_four= Present_values(UK_df, 'United_Kingdom')

#for Argentina
Arg_four = Present_values(Arg_df, 'Argentina')

#for Arab World
Arab_four = Present_values(Arab_df,'Arab_World')
```

```python
world_df = Concat(BOTS_four,Arab_four,EAP_four,Japan_four,China_four,Uk_four,Arg_four)
```
